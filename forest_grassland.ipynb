```python
def get_season(month):
    if month in [12,1,2]:
        return 'Winter'
    elif month in [3,4,5]:
        return 'Summer'
    elif month in [6,7,8]:
        return 'Monsoon'
    else:
        return 'Post-Monsoon'

dfnew['Season'] = dfnew['Month'].apply(get_season)
```
